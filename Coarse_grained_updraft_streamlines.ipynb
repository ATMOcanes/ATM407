## Define grid and depict vectors 

```python
import numpy as np
import matplotlib.pyplot as plt
# %pylab inline

x = np.arange(-10, 11, 0.2)
z = np.arange(0, 6, 1) + 0.2
xx, zz = np.meshgrid(x, z)

u = 1 - np.cos(np.pi *zz/5.)*np.sign(xx) 
w = np.sin(np.pi *zz/5.) * (np.abs(xx) < 0.0001)

fig, ax = plt.subplots(figsize=(15,3))
plt.contourf(x,z,w); plt.title('True narrow updraft w'); plt.ylim((0,5))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,3))
plt.contourf(x,z,u); plt.title('True narrow updraft u'); plt.colorbar()
plt.show()
```

```python
# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(15,3))
plt.quiver(x, z, u, w, pivot='mid')
plt.streamplot(x, z, u, w)
plt.ylim((0,5))
plt.show()
```

```python
x = np.arange(-10, 11, 1)
z = np.arange(0, 5, 1) + 0.5
xx, zz = np.meshgrid(x, z)

u = 1 - np.cos(np.pi *zz/5.)*np.sign(x) 
w = np.sin(np.pi *zz/5.) * (xx == 0)

fig, ax = plt.subplots(figsize=(15,3))
plt.contourf(x,z,w); plt.title('True narrow updraft w')
plt.show()
```
