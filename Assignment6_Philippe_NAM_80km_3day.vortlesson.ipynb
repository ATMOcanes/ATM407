### Background
Recall the spirit of the *vorticity equation*: It eliminated the PGF from the horizontal momentum equations:  

* Horizontal acceleration equation (F=ma):
$$ \frac{du}{dt} = - \frac{\partial \Phi}{\partial x} + fv $$

$$ \frac{dv}{dt} = - \frac{\partial \Phi}{\partial y} - fu $$

* Eliminate $\Phi$ by subtracting the y derivative of the u equation from the x derivative of the v equation: 

$$ \frac{d}{dt}(\zeta_{abs}) = 0 - \zeta_{abs} (\nabla \cdot V)$$

* Recall that $ \zeta_{abs} = \zeta_{rel} + f $, where f is *planetary vorticity*.

* Divergence changes the ***horizontal area A*** of a mass of a blob of air the same way. 

$$ \frac{1}{A} \frac{dA}{dt} = (\nabla \cdot V) $$

$$ \frac{1}{\zeta_{abs}} \frac{d \zeta_{abs}}{dt} = (\nabla \cdot V) $$

* If the air blob's mass is constant, but A increases, then its depth $\delta p$ decreases. 

$$ \frac{1}{\delta p} \frac{d\delta p}{dt} = -(\nabla \cdot V) $$

* But what is this "blob" of air? It is bounded by two material surfaces -- **such as $\theta$ surfaces**. So the equation above refers to *the $\delta p$ for a given $\delta \theta$*. 

* We can therefore eliminate the divergence term: The ratio stays constant, since convergence *amplifies the numerator* but *decays the denominator* by the same fractional amount. 

$$ \frac{d}{dt}(PV) = 0$$

$$ PV = \frac {\zeta_{abs}} {(\delta p / \delta \theta)}$$

(Often people toss in the constant factor of g to turn $\delta p$ (a *weight* of air) into a *mass*.) 


* Except that $\theta$ isn't quiiite conserved... so we need to write 

$$ \frac{d}{dt}(PV) = \zeta_{abs} \frac {\partial}{\partial p}(d\theta/dt) = \zeta_{abs} \frac {\partial}{\partial p}(Q)$$

where Q is our catch-all letter for "diabatic" heating rate (here expressed in $\theta$ units, although that is still K/s or K/d units).

# Illustrating energy of flow with superposition of vortices

### First, define a Rankine vortex in an x,y space centered on 0

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 30
x,y = np.mgrid[-n:n+1, -n:n+1]


#### The flow around a point vortex: decays as 1/r from the central point. 

# First, define radius and angle from x and y 
radius = np.sqrt(x**2 + y**2)
angle  = np.arctan2(y, x)

# Now, create a pure-vortex flow with outer tangential winds of vmax/r and size vsize
vmax = 10
vsize = 1

vtan = np.copy(x)*0. # right sized container 
vtan[radius >= vsize] = vmax /radius[radius >= vsize]
# Make the flow near the center linear in radius (a Rankine vortex) 
vtan[radius < vsize] = vmax*radius[radius < vsize]

# Convert that back to u and v for display 
vvor =  vtan * np.cos(angle)
uvor = -vtan * np.sin(angle)
```

### Plot 2 shifted vortices in 2 colors

```python
fig, ax = plt.subplots(figsize=(7,7))

# Shift the centered vortex to a new position 
from scipy.ndimage.interpolation import shift
udisp1 = shift(uvor, [3,3], mode='wrap')
vdisp1 = shift(vvor, [3,3], mode='wrap') 

# Display the vector field as arrows, CENTERED ON the point they apply to
plt.quiver(x, y, udisp1, vdisp1, pivot='mid', color='blue')

KE1 = np.sum( (udisp1**2+vdisp1**2)/2 )


# Shift the centered vortex to a new position 
from scipy.ndimage.interpolation import shift
udisp2 = shift(uvor, [-3,-3], mode='wrap')
vdisp2 = shift(vvor, [-3,-3], mode='wrap') 

# Display the vector field as arrows, CENTERED ON the point they apply to
plt.quiver(x, y, udisp2, vdisp2, pivot='mid', color='red')

label = 'Far-field flow of two vortices'
plt.title(label)
plt.show()
```

## Plot the net flow (the sum) in purple

```python
# Sum of the two 
fig, ax = plt.subplots(figsize=(7,7))

# Display the vector field as arrows, CENTERED ON the point they apply to
plt.quiver(x, y, udisp1+udisp2, vdisp1+vdisp2, pivot='mid', color='purple')

label = 'Far-field flow of two vortices'
plt.title(label)
plt.show()
```

## Kinetic energy as a function of distance between the vortices

```python
half_distances = np.arange(10)
KEs = np.copy(half_distances)*0. # right sized container for answers

for index,dist in enumerate(half_distances): 
    fig, ax = plt.subplots(figsize=(7,7))

# Shift each centered vortex to its new position 
    udisp1 = shift(uvor, dist, mode='wrap')
    vdisp1 = shift(vvor, dist, mode='wrap') 
    udisp2 = shift(uvor, -dist, mode='wrap')
    vdisp2 = shift(vvor, -dist, mode='wrap') 

    u = udisp1+udisp2
    v = vdisp1+vdisp2
    KE = np.sum( (u**2+v**2)/2 )/KE1
    KEs[index] = KE
    
# Display the vector field as arrows, CENTERED ON the point they apply to
    plt.quiver(x, y, u, v, pivot='mid')

    label = 'Two vortices: KE= '+str(KE)+' times that of a single vortex'
    plt.title(label)
    plt.show()
```
