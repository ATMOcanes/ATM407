### Background
Recall the spirit of the *vorticity equation*: It eliminates the PGF from the horizontal momentum equations.  Let's ignore friction for now.  

* Horizontal acceleration equation (F=ma):
$$ \frac{du}{dt} = - \frac{\partial \Phi}{\partial x} + fv $$

$$ \frac{dv}{dt} = - \frac{\partial \Phi}{\partial y} - fu $$

* Eliminate $\Phi$ by subtracting the y derivative of the u equation from the x derivative of the v equation: 

$$ \frac{d}{dt}(\zeta_{abs}) = 0 - \zeta_{abs} (\nabla \cdot V)$$

* Recall that $ \zeta_{abs} = \zeta_{rel} + f $, where f is *planetary vorticity*.

* We neglect the divergence term, or view it (vaguely) as one aspect of the "forcing" of Rossby waves. 

$$ \frac{d}{dt}(\zeta_{abs}) = 0 $$



* In our Irma-related exercise, we focused on the advection of $\zeta_{rel}$ by winds circulating around patches of $\zeta_{rel}$. 

* Here we instead consider the *Rossby wave* limiting case: considering **ONLY** the horizontal advection of planetary vorticity, again by winds $\vec V_{hor}$ circulating around patches of $\zeta_{rel}$. 

$$ \frac{\partial}{\partial t}(\zeta_{rel}) = -\vec V_{hor} \cdot \nabla f$$
