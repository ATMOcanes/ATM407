### Background: Nie and Fan 2019

[Nie and Fan](https://journals.ametsoc.org/doi/full/10.1175/JCLI-D-19-0188.1) describe their inversion of Quasi-Geostrophic omega equation from ERA Interim data. These are their results, stored at a [Miami data server](https://weather.rsmas.miami.edu/repository/entry/show?entryid=1f1f6f3f-d1e2-45b4-ad25-d4f6b8873b4a). This notebook shows only the **composite mean over the 39 years** (heaviest rain event in each summer).
